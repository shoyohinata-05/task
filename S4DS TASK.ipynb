```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv(r"C:\Users\91986\Downloads\E-Commerce_Data\Amazon_Products.csv")
data_set
```

```python
dataset = data_set[['uniq_id','product_name','price','number_available_in_stock','average_review_rating','number_of_reviews']].copy()
dataset
```

```python
dataset['price'] = dataset['price'].str.replace(r'£', '')
dataset['average_review_rating'] = dataset['average_review_rating'].str.replace(r'out of 5 stars', '')
dataset['number_available_in_stock'] = dataset['number_available_in_stock'].str.replace(r'new', '')
dataset
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(axis=0, how='any',  subset='price', inplace=False)
dataset
```

```python
df=dataset.drop(['product_name'], axis=1)
```

```python
df['number_available_in_stock'] = df['number_available_in_stock'].fillna(0)
df
```

```python
df["price"] = pd.to_numeric(df["price"], downcast="float",errors='ignore')
print(df.dtypes)
```

```python
df['price'] = pd.to_numeric(df['price'],errors = 'coerce')
df['average_review_rating'] = pd.to_numeric(df['average_review_rating'],errors = 'coerce')
df['number_of_reviews'] = pd.to_numeric(df['number_of_reviews'],errors = 'coerce')
print(df.info())
```

```python
sns.scatterplot(x = 'price',y = 'average_review_rating', data = df)
```
